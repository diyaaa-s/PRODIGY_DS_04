```python
# ✅ Import necessary libraries (JupyterLite supported)
import pandas as pd
import matplotlib.pyplot as plt

# ✅ Load the dataset
df = pd.read_csv("twitter_training.csv", header=None, names=["Tweet_ID", "Entity", "Sentiment", "Tweet"])
df.dropna(inplace=True)

# ✅ Show dataset shape and a few rows
print("Dataset Shape:", df.shape)
print(df[['Sentiment', 'Tweet']].head())

# ✅ Count the number of tweets per sentiment
sentiment_counts = df['Sentiment'].value_counts()

# ✅ Bar plot for sentiment distribution
plt.figure(figsize=(6, 4))
plt.bar(sentiment_counts.index, sentiment_counts.values, color=['green', 'blue', 'red'])
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Tweet Count")
plt.show()

# ✅ Count the most mentioned entities
top_entities = df['Entity'].value_counts().head(10)

# ✅ Bar plot for top 10 entities
plt.figure(figsize=(8, 5))
plt.bar(top_entities.index, top_entities.values, color='orange')
plt.title("Top 10 Most Mentioned Entities")
plt.xlabel("Entity")
plt.ylabel("Mentions")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
